# Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Load Dataset

```python
# Ganti dengan nama file dataset kamu
df = pd.read_csv("btc_15m_data_2018_to_2025.csv")

# Lihat 5 baris teratas
df.head()
```

# Cek Nama Kolom

```python
df.columns
```

# Konversi Kolom Waktu & Set Index

```python
# Konversi kolom waktu menjadi datetime dan set sebagai index
df['Open time'] = pd.to_datetime(df['Open time'])
df.set_index('Open time', inplace=True)

# Tampilkan 5 baris teratas untuk memastikan
df.head()
```

# Visualisasi Plot Time Series

```python
plt.figure(figsize=(12, 6))
df['Close'].plot()
plt.title('Pergerakan Harga Penutupan Bitcoin (20182025)')
plt.xlabel('Tahun')
plt.ylabel('Harga Bitcoin (USD)')
plt.grid(True)
plt.show()
```
